```python
total_miles = [11.4,8.7,5.5,7,7.7,9.1,7.2,1.3,9.6,13.9,6.2,5.6,7.6,6.9,7.1,7.3,8.4,7,9.1,5.8,5.4,6.1,6,3.3]
walked_miles = [4.5,5.4,3.1,4.0,3.2,3.5,4.4,0.0,6.4,5.1,4.1,3.5,3.5,4.6,2.5,6.1,5.6,3.9,5.4]
diff = [total_miles[i] - walked_miles[i] for i in range(len(walked_miles))]
#print(diff)

avg = sum(diff)/len(diff)

print(avg)

walked_miles_pred = [total_miles[i] - avg for i in range(len(total_miles))]
print(walked_miles_pred)
```
