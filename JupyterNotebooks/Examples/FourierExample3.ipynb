```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi,np.pi,1000)
```

```python
fx = []
for i in range(len(x)):
    fx.append(np.exp(x[i]))
```

```python
plt.plot(x,fx)
```

```python
x2 = np.linspace(-np.pi,np.pi,1000)
fx_fourier = np.zeros(len(x2))

a0 = 1/np.pi*(np.exp(np.pi)-np.exp(-np.pi))

nterms = 4

for i in range(nterms+1):
    for j in range(len(x2)):
        if i==0:
            fx_fourier[j] = fx_fourier[j]+a0/2
        else:
            an = (-1)**i*(np.exp(np.pi)-np.exp(-np.pi))/(np.pi*(1+i**2))
            bn = -i*an
            if (j==0 and i<=10): 
                print ("n = %0.3f, a_n = %0.3f, bn = %0.3f" % (i,an,bn))
            fx_fourier[j] = fx_fourier[j]+an*np.cos(i*x2[j])+bn*np.sin(i*x2[j])
```

```python
plt.plot(x,fx)
plt.plot(x2,fx_fourier)
plt.show()
```
