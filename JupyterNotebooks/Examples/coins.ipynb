```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a random list of 100 numbers, either 0 or 1
np.random.seed(1131)
n = 100
```

```python
total1_stats = []
total2_stats = []
difference = []
a_wins = 0
b_wins = 0
ties = 0

ntrials = 1000000

for iter in range(ntrials):
    # generate a random list of 0's and 1's
    x = np.random.randint(0, 2, n) # 0 or 1
    
    # search through the list one at a time, until we find a two 1's in total

    # initialize the variables
    count = 0
    total1 = 0
    # loop through the list
    for i in range(n):
        count += x[i]
        total1 += 1
        if count == 2:
            break
        
    #print(total1)
    total1_stats.append(total1)

    # Now, search through the list looking at every other element, until we find two 1's in total

    # initialize the variables
    count = 0
    total2 = 0
    # loop through the list
    for i in range(1, n, 2):
        count += x[i]
        total2 += 1
        if count == 2:
            break
        
    #print(total2)
    total2_stats.append(total2)
    
    difference.append(total1 - total2) 
    # difference between the two totals - 
    # bob wins if total1 > total2, alice wins if total1 < total2, neither wins if total1 = total2
    
    if total1 < total2:
        a_wins += 1
    elif total1 > total2:
        b_wins += 1
    else:
        ties += 1
        
print('alice wins:', a_wins/ntrials*100.0, '%')
print('bob wins:', b_wins/ntrials*100.0, '%')
print('alice and bob tie:', ties/ntrials*100.0, '%')
```

```python
# plot the histograms

plt.hist(total1_stats, bins=range(1, 21), alpha=0.5, label='total1')
plt.hist(total2_stats, bins=range(1, 21), alpha=0.5, label='total2')
plt.legend(loc='upper right')
plt.show()

# calculate the mean and standard deviation of the two lists
print('mean of total1:', np.mean(total1_stats))
print('std error in mean of total1:', np.std(total1_stats)/np.sqrt(ntrials))
print('mean of total2:', np.mean(total2_stats))
print('std error in mean of total2:', np.std(total2_stats)/np.sqrt(ntrials))
```

```python
plt.hist(difference, bins=range(-10, 11), alpha=0.5, label='difference')
plt.legend(loc='upper right')

# calculate the mean and standard deviation of the difference
print('mean of difference:', np.mean(difference))
print('std error in mean of difference:', np.std(difference)/np.sqrt(ntrials))
```
