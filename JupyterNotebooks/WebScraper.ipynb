```python
import requests

url = 'https://cnu.edu/public/undergradcatalog/_pdf/cnu-undergraduate_catalog_20-21.pdf'

response = requests.get(url)

with open('/tmp/metadata.pdf', 'wb') as f:
    print ('Writing content to /tmp/metadata.pdf')
    f.write(response.content)
```

```python
# importing required modules 
import PyPDF2
  
# creating a pdf file object 
pdfFileObj = open('/tmp/metadata.pdf', 'rb') 
  
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
  
# printing number of pages in pdf file 
print(pdfReader.numPages) 
print()
  
# creating a page object 
pageObj = pdfReader.getPage(0) 
  
# extracting text from page 
print(pageObj.extractText()) 
  
# closing the pdf file object 
pdfFileObj.close() 
```

```python
def print_out(term_code_number,crn_number):
    
    first_part = "https://navigator.cnu.edu/FacultyScheduleofClasses/socadditional.aspx?term_code="
    next_part = "&crn="
    
    my_url = first_part + str(term_code_number) + next_part + str(crn_number)
    #print (my_url)
    
    page = urlopen(my_url)
    html = page.read().decode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    
    for element in soup.find_all('span'):
        try:
            
            if element.attrs['id'] == 'FormView1_CRNLabel':
                course_CRN = element.get_text()
                #print('------------------')
                print ('CRN:', course_CRN)
                #print()
            
            if element.attrs['id'] == 'FormView1_SUBJ_CODELabel':
                course_prefix = element.get_text()
        
            if element.attrs['id'] == 'FormView1_CRSE_NUMBLabel':
                course_number = element.get_text()
                #print (course_prefix,course_number)
                #print()  
            
            if element.attrs['id'] == 'FormView1_COURSE_DESCLabel':
                course_desc = element.get_text()
                #print (course_desc)
     
        except:
            continue
            
    try:    
        return course_CRN,course_prefix,course_number,course_desc
    except:
        return 0,0,0,'empty'
        
```

```python
crn,prefix,number,desc = print_out(202110,8165)
```

```python
crn = []
prefix = []
number = []
desc = []

pold = 'AAAA'
nold = 0

for i in range(8000,10000):
    c,p,n,d = print_out(202110,i)
    #print (p,pold,n,nold)
    if c != 0 and (p != pold or (p == pold and n != nold)):
        crn.append(c)
        prefix.append(p)
        number.append(n)
        desc.append(d)
        pold = p
        nold = n
        
df = pd.DataFrame({'CRN':crn, 'Prefix':prefix, 'Number':number, 'Description':desc})
```
