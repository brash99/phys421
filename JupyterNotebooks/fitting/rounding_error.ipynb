 Set up arrays to hold values, and choose plotting options.

```python
from array import array
import math

ichoice = 3

n = 20

# derivative functions:  exact, using forward derivative approx., and using +/- derivative approx
dfunc = array('d')
dfunch = array('d')
dfunchh = array('d')

# basic functions:  f(x), f(x+h), f(x-h), f(x+2h), f(x-2h)
xfunc = array('d')
xfunchp = array('d')
xfunchm = array('d')

# h values
x = array('d')

# difference arrays for (a) forward derivative and (b) +/- derivative 
diff = array('d')
difftwo = array('d')

xval = 1.0

if ichoice == 1:
    print ("Using forward derivative ... ")
elif ichoice == 2:
    print ("Using centered derivative ... ")
else:
    print ("Using both methods ... ")
```

Define the function and its derivative

```python
def myfunction(x):
    return x*x*x
```

```python
def myderivfunction(x):
    return 3.0*x*x
```

Loop over h values:  calculate derivative functions exactly, and using approximations.

```python
for i in range(0,n):

    hpower=i-n;
    x.append(math.pow(10.0,int(hpower)));
    
    xp = xval+x[i]
    xm = xval-x[i]
    
    xfunc.append(myfunction(xval));
    xfunchp.append(myfunction(xp));
    xfunchm.append(myfunction(xm));
    dfunc.append(myderivfunction(xval));
    
    if ichoice == 1:
        dfunch.append((xfunchp[i]-xfunc[i])/(x[i]));
        diff.append(abs(dfunc[i]-dfunch[i]));
    elif ichoice == 2:
        dfunch.append((xfunchp[i]-xfunchm[i])/(2.0e+00*x[i]));
        diff.append(abs(dfunc[i]-dfunch[i]));
    else:
        dfunch.append((xfunchp[i]-xfunc[i])/(x[i]));
        dfunchh.append((xfunchp[i]-xfunchm[i])/(2.0e+00*x[i]));
        diff.append(abs(dfunc[i]-dfunch[i]));
        difftwo.append(abs(dfunc[i]-dfunchh[i]));
```

Plot the results!

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the first graph
ax.plot(x, diff, 'b-o', markersize=5, label='Forward Difference')

# Set title and labels for the first graph
ax.set_title('Rounding Error')
ax.set_xlabel('X')
ax.set_ylabel('Error')

# Plot the second graph
ax.plot(x, difftwo, 'r-o', markersize=5, label='Central Diffence')

# Set title and labels for the second graph
ax.set_title('Rounding Error')
ax.set_xlabel('X')
ax.set_ylabel('Error')

# Set logarithmic scale for both x and y axes
ax.set_xscale('log')
ax.set_yscale('log')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
